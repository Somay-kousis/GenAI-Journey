# Models

## 1. Language Models
Language Models can be broadly categorized into:

- **LLMs (Large Language Models)**
- **Chat Models**

### A. LLMs (Large Language Models)
General-purpose language models designed for a wide range of NLP tasks.

**Common Use Cases:**
- Text generation
- Summarization
- Question answering
- Translation
- Content creation

**Examples:**
- GPT-3
- LLaMA
- Mistral

### B. Chat Models
Language models optimized for conversational interactions.

**Common Use Cases:**
- Chatbots
- Virtual assistants
- Customer support
- Multi-turn conversations
- Agent-based workflows

**Examples:**
- ChatGPT
- Claude
- Gemini

> **Note:** Modern AI systems increasingly use Chat Models because they combine the capabilities of traditional LLMs with instruction-following, conversation management, and tool usage.

---

## 2. Embedding Models

Embedding models convert text, images, or other data into numerical vector representations (embeddings).

**Common Use Cases:**
- Semantic search
- Retrieval-Augmented Generation (RAG)
- Recommendation systems
- Clustering
- Similarity matching

**Examples:**
- text-embedding-3-large
- text-embedding-3-small
- BGE Embeddings
- E5 Embeddings

---

# Source Availability

Both Language Models and Embedding Models can be:

## Closed Source
- Model weights are not publicly available.
- Access is typically through APIs.

**Examples:**
- ChatGPT
- Claude
- Gemini

## Open Source
- Model weights are publicly available.
- Can be downloaded, fine-tuned, and self-hosted.

**Examples:**
- LLaMA
- Mistral
- Gemma
- Qwen
